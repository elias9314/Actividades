```python
# Ejercicios de refuerzo
# Hacer una función que saque el mínimo común multiplo de dos números
num1=int(input("digite el primer número: "))
num2=int(input("digite el segundo número: "))
def minimo(num1,num2):
    result = max(num1,num2)
    while True:
        if (result%num1==0 and result%num2==0):
            return result
        result += 1
        
minimo(num1,num2)
```

```python
# Hacer una función que saque el máximo común divisor de dos números
num1=int(input("digite el primer número: "))
num2=int(input("digite el segundo número: "))
def maximo(num1,num2):
    mcd=False
    if(num1>0 and num2>0 and num1!=num2):
        if(num2<num1):
            espacio = num1
            num1 = num2
            num2 = espacio
        i = num1
        while not mcd and i>=1:
            if(num1 % i==0 and num2 % i==0):
                print("El máximo común divisor es :",i)
                mcd= True
            else:
                i-=1
    else:
        print("Los números digitados son incorrectos")
maximo(num1,num2)
```
